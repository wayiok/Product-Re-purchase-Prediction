```python
train_dataframes = []
for i in tqdm(range(1, 3)):
    train_dataframes.append(pd.read_csv(f'{trainFolder}/train_data_part_{i}.csv'))
train_data = pd.concat(train_dataframes, ignore_index=True)
del train_dataframes
```

```python
merged_df = train_data.copy()
merged_df['date'] = pd.to_datetime(merged_df['date'])
merged_df['year'] = merged_df['date'].dt.year
merged_df['month'] = merged_df['date'].dt.month
##Frecuency by month of the year
del train_data
frecuency = merged_df.groupby(['customer_id', 'product_id','year','month'])['transaction_id'].transform('count')
merged_df['frecuency'] = frecuency
merged_df.sample(4)
```

```python
merged_df = merged_df[['year','month', 'transaction_id', 'customer_id', 'product_id', 'frecuency']]
merged_df.sample(4)
```

```python
# prompt: by date, transaction_id , customer_id change product_id and frecuency in to list.
del products_data
grouped_df = merged_df.groupby(['year','month', 'transaction_id', 'customer_id']) \
    .agg({'product_id': list, 'frecuency': list}) \
    .reset_index()

grouped_df
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming `grouped_df` is your DataFrame with `product_id` column as lists of products in each transaction
# Sample data structure:
# grouped_df = pd.DataFrame({
#     'transaction_id': ['Transaction_1', 'Transaction_2'],
#     'product_id': [['Product_A', 'Product_B'], ['Product_B', 'Product_C', 'Product_D']]
# })

# Step 1: Transform Data
# Convert the `product_id` column into a list of lists for each transaction
transactions = grouped_df['product_id'].tolist()

# Step 2: One-Hot Encode Transaction Data
# This transforms each product in a transaction into a column, with a binary 1 or 0 indicating if the product was bought.
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Step 3: Apply Apriori to Find Frequent Itemsets
# We set a minimum support threshold (adjust as needed)
min_support = 0.01  # 1% support threshold
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

# Step 4: Generate Association Rules
# Set minimum confidence threshold (adjust as needed)
min_confidence = 0.1  # 30% confidence threshold
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

# Step 5: Analyze the Rules
# Sort rules by lift (higher lift indicates stronger association)
rules = rules.sort_values(by="lift", ascending=False)

# Display the results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
import pandas as pd
from collections import Counter

# Assuming 'rules' DataFrame is from the previous association rules code
# and 'grouped_df' contains customer transactions with columns 'customer_id' and 'product_id'.

# Step 1: Calculate top 10 most frequently purchased products as fallback
all_products = [product for sublist in grouped_df['product_id'] for product in sublist]
top_products = [product for product, _ in Counter(all_products).most_common(10)]

# Step 2: Create a dictionary to store predictions for each customer
customer_predictions = {}

# Go through each customer in the original data
for _, row in grouped_df.iterrows():
    customer_id = row['customer_id']
    purchased_products = set(row['product_id'])

    # Collect potential products and their scores for the customer
    predictions = []
    for _, rule in rules.iterrows():
        antecedent = rule['antecedents']
        consequent = rule['consequents']

        # Check if customer has bought the antecedent products and not the consequent products
        if antecedent.issubset(purchased_products) and not consequent.issubset(purchased_products):
            predictions.append((list(consequent)[0], rule['confidence']))

    # Sort predictions by confidence score in descending order and keep unique products
    predictions = sorted(predictions, key=lambda x: x[1], reverse=True)
    predicted_products = list({p[0] for p in predictions[:10]})  # Get unique top predictions

    # Ensure we have exactly 10 products by filling up with top products if needed
    if len(predicted_products) < 10:
        predicted_products += [p for p in top_products if p not in predicted_products][:10 - len(predicted_products)]

    # Rank from 10 to 1 (highest to lowest)
    ranks = list(range(10, 0, -1))

    # Store the final predictions for the customer
    customer_predictions[customer_id] = {
        'product_id': predicted_products,
        'rank': ranks
    }

# Step 3: Convert the predictions dictionary to a DataFrame
predictions_df = pd.DataFrame([
    {'customer_id': customer, 'product_id': data['product_id'], 'rank': data['rank']}
    for customer, data in customer_predictions.items()
])

# Display the resulting DataFrame
print(predictions_df)

```
