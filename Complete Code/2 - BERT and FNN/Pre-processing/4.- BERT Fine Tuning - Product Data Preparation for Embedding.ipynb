
### Full text generation

```python
from sklearn.preprocessing import LabelEncoder

# Assume 'category' is the column with string labels
label_encoder = LabelEncoder()
products_data['category'] = products_data['shelf_level1']
products_data['shelf_level1'] = label_encoder.fit_transform(products_data['shelf_level1'])
```

```python
# List of binary columns in products_data
binary_columns = [
    'bio', 'sugar_free', 'aspartame_free', 'gluten_free', 'halal', 'casher', 'eco_friendly',
    'local_french', 'artificial_coloring_free', 'taste_enhancer_free', 'antibiotic_free',
    'reduced_sugar', 'vegetarian', 'pesticide_free', 'grain_free', 'no_added_sugar',
    'salt_reduced', 'nitrite_free', 'fed_without_ogm', 'no_added_salt', 'no_artificial_flavours',
    'porc', 'vegan', 'frozen', 'fat_free', 'reduced_fats', 'fresh', 'alcool',
    'lactose_free', 'phenylalanine_free', 'palm_oil_free', 'produits_du_monde',
    'regional_product', 'national_brand', 'first_price_brand', 'carrefour_brand'
]

# Replace missing values with placeholders
products_data.fillna('unknown', inplace=True)

# Updated lists of relevant features
relevant_positive = ['bio', 'sugar_free', 'gluten_free', 'vegan', 'national_brand', 'regional_product', 'first_price_brand', 'carrefour_brand']
relevant_negative = ['bio', 'sugar_free', 'gluten_free', 'palm_oil_free', 'alcool', 'carrefour_brand']

# Function to prioritize and simplify binary features
def summarize_binary_features(row):
    positive = ['This product ']
    negative = [' [SEP] This product ']

    # Process positive attributes
    for feature in relevant_positive:
        if row[feature] == 1:
            positive.append(f"is {feature.replace('_', ' ')}")

    # Process negative attributes
    for feature in relevant_negative:
        if row[feature] == 0:
            negative.append(f"is not {feature.replace('_', ' ')}")

    # Combine into a single text
    return ', '.join(positive) + ' , '.join(negative)

# Apply the function to create binary features text
products_data['binary_features_text'] = products_data.apply(summarize_binary_features, axis=1)
products_data['binary_features_text'] = products_data['binary_features_text'].str.rstrip('.')
```

```python
# Create final combined text
products_data['text_for_bert'] = (
    'Description: '+ products_data['product_description'].str.lower() + ' [SEP] ' +
    'Brand: ' + products_data['brand_key'] + ' [SEP] ' +
    'Sector: ' + products_data['sector'] + ' [SEP] ' +
    'Category: ' + products_data['category'] + ' [SEP] ' +
    'Subcategory: ' + products_data['shelf_level2'] + ' [SEP] '  +
    products_data['binary_features_text'] + ' [SEP] ' +
    'Product Id: ' + products_data['product_id']
)

# Preview the final text
products_data[['product_id', 'text_for_bert','shelf_level1']].head()
```

### Fine Tuning

```python
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset, DataLoader
import torch
```

```python

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
#model = BertModel.from_pretrained('bert-base-uncased') # Original code
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(products_data['shelf_level1'].unique())) # Updated model loading
```

```python
model.to('cuda')
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=10,
    per_device_train_batch_size=16,
    learning_rate=2e-5,
    fp16=True,  # Enable mixed precision training
    logging_dir='./logs',
    save_steps=500,
)
```

```python
# Define Dataset
class ProductDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_len):
        self.tokenizer = tokenizer
        self.data = dataframe
        self.max_len = max_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        product_info = self.data.iloc[index]
        full_text = product_info['text_for_bert']

        inputs = self.tokenizer.encode_plus(
            full_text,
            None,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            return_token_type_ids=True,
            truncation=True
        )
        label = product_info['shelf_level1']  # Replace with the actual label column

        return {
            'input_ids': torch.tensor(inputs['input_ids'], dtype=torch.long),
            'attention_mask': torch.tensor(inputs['attention_mask'], dtype=torch.long),
            'labels': torch.tensor(label, dtype=torch.long)
        }
```

```python
product_dataset = ProductDataset(products_data, tokenizer, max_len=128)
```

```python
# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=product_dataset,
)

# Fine-tune the model
trainer.train()
```

```python
# Load the fine-tuned model for embedding extraction
model = BertModel.from_pretrained('./results/checkpoint-51500')
model.to('cuda')  # Move the model to GPU

# Function to get product embedding
def get_product_embedding(product_text):
    # Tokenize the input text and move the tensors to GPU
    inputs = tokenizer(product_text, return_tensors='pt').to('cuda')

    # Put the model in evaluation mode to disable dropout layers
    model.eval()

    # Pass the inputs through the model and move the output to CPU to detach
    with torch.no_grad():
        outputs = model(**inputs)

    # Extract the CLS token representation as embedding
    cls_embedding = outputs.last_hidden_state[:, 0, :].cpu().numpy()
    return cls_embedding
```

```python
from tqdm import tqdm

# Generate embeddings for all products with progress indication
product_embeddings = []
for i, row in tqdm(products_data.iterrows(), total=len(products_data), desc="Generating Product Embeddings"):
    product_text = row['text_for_bert']
    embedding = get_product_embedding(product_text)
    product_embeddings.append((row['product_id'], embedding))

# Example of storing embeddings as a NumPy array
product_embeddings_np = np.array([embedding for _, embedding in product_embeddings])
np.save(f'{trainFolder}/product_embeddings_2.npy', product_embeddings_np)

# Save product IDs separately
product_ids = [product_id for product_id, _ in product_embeddings]
np.save(f'{trainFolder}/product_ids_2.npy', product_ids)
```
