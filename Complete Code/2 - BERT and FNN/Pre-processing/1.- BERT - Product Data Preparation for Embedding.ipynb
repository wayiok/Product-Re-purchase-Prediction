```python
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd

# Initialize BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')
```

```python
products_data['sector'] = products_data['sector'].map({
    'PGC': 'Fast-Moving Consumer Goods',
    'PRODUITS FRAIS TRANSFORMATION': 'Fresh product processing',
    'BAZAR': 'Bazar',
    'EPCS': 'Electro Photo Cine Sound',
    'TEXTILE': 'textile',
    'ACTIVITES PERIPHERIQUES': 'Peripheral Activities',
    })
```

```python
shelf1_values = products_data['shelf_level1'].unique()
shelf2_values = products_data['shelf_level2'].unique()

translation_dict = {
    'shelf_level1': shelf1_values,
    'shelf_level2': shelf2_values,
}
translation_dict
```

```python
from googletrans import Translator

translator = Translator()

new_translation_dict = {}
for key, values in translation_dict.items():
  new_translation_dict[key] = {}
  for value in values:
    try:
      translated_value = translator.translate(value, dest='en').text
      new_translation_dict[key][value] = translated_value
    except Exception as e:
      print(f"Error translating {value}: {e}")
      new_translation_dict[key][value] = None

new_translation_dict
```

```python
# prompt: now change the values in products_data of columns "shelf_level1" and "shelf_level2" wiht this dictionary.

for col in ['shelf_level1', 'shelf_level2']:
  products_data[col] = products_data[col].map(new_translation_dict[col])
```

```python
# List of binary columns in products_data
binary_columns = [
    'bio', 'sugar_free', 'aspartame_free', 'gluten_free', 'halal', 'casher', 'eco_friendly',
    'local_french', 'artificial_coloring_free', 'taste_enhancer_free', 'antibiotic_free',
    'reduced_sugar', 'vegetarian', 'pesticide_free', 'grain_free', 'no_added_sugar',
    'salt_reduced', 'nitrite_free', 'fed_without_ogm', 'no_added_salt', 'no_artificial_flavours',
    'porc', 'vegan', 'frozen', 'fat_free', 'reduced_fats', 'fresh', 'alcool',
    'lactose_free', 'phenylalanine_free', 'palm_oil_free', 'produits_du_monde',
    'regional_product', 'national_brand', 'first_price_brand', 'carrefour_brand'
]

# Replace missing values with placeholders
products_data.fillna('unknown', inplace=True)

# Updated lists of relevant features
relevant_positive = ['bio', 'sugar_free', 'vegan', 'national_brand', 'regional_product', 'first_price_brand','halal', 'carrefour_brand']
relevant_negative = ['bio', 'sugar_free', 'gluten_free', 'artificial_coloring_free', 'taste_enhancer_free', 'palm_oil_free', 'alcool', 'carrefour_brand']

# Function to prioritize and simplify binary features
def summarize_binary_features(row):
    positive = []
    negative = []

    # Process positive attributes
    for feature in relevant_positive:
        if row[feature] == 1:
            positive.append(f"is {feature.replace('_', ' ')}")

    # Process negative attributes
    for feature in relevant_negative:
        if row[feature] == 0:
            negative.append(f"is not {feature.replace('_', ' ')}")

    # Combine into a single text
    return ', '.join(positive) + '. ' + ', '.join(negative)

# Apply the function to create binary features text
products_data['binary_features_text'] = products_data.apply(summarize_binary_features, axis=1)
products_data['binary_features_text'] = products_data['binary_features_text'].str.rstrip('.').str.lower()

# Create final combined text
products_data['text_for_bert'] = (
    'The product with the description: ' + products_data['product_description'].str.lower() +
    ' from the ' + products_data['brand_key'] +
    ' brand, is under the ' + products_data['sector'] + ' sector. ' +
    'It belongs to the category: ' + products_data['shelf_level1'] +
    ' and is further classified as: ' + products_data['shelf_level2'] + '. ' +
    products_data['binary_features_text']
)

# Preview the final text
products_data[['product_id', 'text_for_bert']].head()

```

```python
# Function to generate BERT embeddings
def get_bert_embedding(text):
    # Tokenize and encode text
    inputs = tokenizer(text, return_tensors='pt', truncation=True, max_length=512, padding='max_length')
    outputs = bert_model(**inputs)
    # Use the CLS token embedding as the representation
    cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze().detach().numpy()
    return cls_embedding

# Apply BERT embedding generation to all product text
print("Generating embeddings for products...")
products_data['embedding'] = products_data['text_for_bert'].apply(get_bert_embedding)

# Verify embeddings
print("Sample embedding for a product:", products_data['embedding'].iloc[0])
```
